This is to manually update each annotation in the toy dataset v0.0.1 and standardized dataset v0.1.0, especially to include longitude and latitude.

```python
from wsrdata.utils.geo_util import get_roost_coor
import json

JSON_PATHS = [
    "../datasets/roosts_v0.0.1/roosts_v0.0.1.json",
    "../datasets/roosts_v0.1.0/roosts_v0.1.0.json"
]
geosize = 300000 # 300km
imsize = 600 # 600 pixels
station_xy = (imsize / 2., imsize / 2.)
distance_per_pixel = geosize / imsize

for json_path in JSON_PATHS:
    if json_path == "../datasets/roosts_v0.0.1/roosts_v0.0.1.json":
        new_json_path = "../datasets/roosts_v0.0.1_official/roosts_v0.0.1.json"
        pretty_print_indent = 4
    else:
        new_json_path = "../datasets/roosts_v0.1.0_official/roosts_v0.1.0.json"
        pretty_print_indent = None
    
    with open(json_path, "r") as f:
        dataset = json.load(f)
        dataset_version = dataset["info"]["dataset_version"]
        
    annotations = dataset["annotations"]
    dataset["annotations"] = []
    
    for i, annotation in enumerate(annotations):
        if i % 200 == 0:
            print(i)
        station_name = dataset["scans"][annotation["scan_id"]]["key"][:4]
        roost_xy = [annotation["x_im"], annotation["y_im"]]
        lon, lat = get_roost_coor(
            roost_xy, 
            station_xy, 
            station_name, 
            distance_per_pixel,
            y_direction="geographic"
        )
        
        dataset["annotations"].append({
            "id": annotation["id"],
            "scan_id": annotation["scan_id"],
            "category_id": annotation["category_id"],
            "dataset_version": dataset_version,
            "track_id": dataset_version + "-" + str(annotation["sequence_id"]),
            "lon": lon,
            "lat": lat,
            "radius": annotation["r_im"] * distance_per_pixel,
            "x_im": annotation["x_im"],
            "y_im": annotation["y_im"],
            "r_im": annotation["r_im"],
            "bbox": annotation["bbox"],
            "bbox_area": annotation["bbox_area"],
            "subcategory": None,
            "notes": None,
            "day_notes": None
        })
        
    with open(new_json_path, 'w') as f:
        json.dump(dataset, f, indent=pretty_print_indent)
```
